```
df1=df.drop(["Province/State","Lat","Long"],axis=1)
```

```
df2=df1.isin([0]).sum()
```

```
col=[]
x=len(df)*0.7
for i in range(1,62):
 if(df2[i]>x):
    col.append(i)

print(col)    
df1.drop(df1.columns[col],axis=1,inplace=True)
    
df1


```

```
df1.drop(["Country/Region"],axis=1, inplace=True)
```

```
sum_s=df1.sum(axis=0)
sum_s
```

```
import matplotlib.pyplot as plt
plt.figure(figsize=(11,12))
plt.pie(sum_s,labels=df1.columns,labeldistance=1.2,autopct='%.3f%%')
plt.title('After Column reduction\n\n',size=30) 
plt.legend(df1.columns,loc=10, prop={'size':8})
```
